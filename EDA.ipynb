```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Path to the directories
normal_dir = "data/train/NORMAL/"
pneumonia_dir = "data/train/PNEUMONIA/"

# Get the list of all files in each directory
normal_files = os.listdir(normal_dir)
pneumonia_files = os.listdir(pneumonia_dir)

# Create a DataFrame to store the file names and labels
data = {
    'filename': normal_files + pneumonia_files,
    'label': ['NORMAL'] * len(normal_files) + ['PNEUMONIA'] * len(pneumonia_files)
}
df = pd.DataFrame(data)
```

```python
# Plot the distribution of the classes
df['label'].value_counts().plot(kind='bar')
plt.title("Class distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
# Plot a few random images
random_indices = np.random.choice(df.index, size=5, replace=False)

for i, idx in enumerate(random_indices):
    file = df.iloc[idx]['filename']
    label = df.iloc[idx]['label']
    if label == 'NORMAL':
        path = normal_dir + file
    else:
        path = pneumonia_dir + file
    plt.subplot(1, 5, i+1)
    plt.imshow(plt.imread(path), cmap='gray')
    plt.title(label)
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Plot the distribution of image shapes
heights = []
widths = []

for file in df['filename']:
    path = normal_dir + file if file in normal_files else pneumonia_dir + file
    image = plt.imread(path)
    heights.append(image.shape[0])
    widths.append(image.shape[1])

plt.hist(heights, bins=50)
plt.title("Distribution of image heights")
plt.xlabel("Height")
plt.ylabel("Count")
plt.show()

plt.hist(widths, bins=50)
plt.title("Distribution of image widths")
plt.xlabel("Width")
plt.ylabel("Count")
plt.show()
```
